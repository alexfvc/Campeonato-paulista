#### LIMPEZA 

```python
colunasjogos = dfjogos.columns.to_list()
colunasjogos
```

```python
dfjogos = dfjogos.drop(labels=[ 
    'eventStatus', 'eventLeague','underOver', 'predGoalsHome', 
    'predGoalsAway', 'advice', 'formHome', 
    'formAway', 'attHome', 'attAway', 'defHome', 
    'defAway', 'poissonHome', 'poissonAway', 'h2hHome', 
    'h2hAway', 'goalsHome', 'goalsAway', 'totalHome', 
    'totalAway', 'homeOdd', 'drawOdd', 'awayOdd', 'over05', 
    'over05Ht', 'under15', 'under15Ht','date','hour','winner',
    'winOrDraw','logoAwayTeam','logoHomeTeam'],axis=1) 
#drop colunas que nao rerao usadas
```

```python
dfstatus = dfstatus.replace(np.nan,0.0) #renomeia Nan para 0.0
dfjogos = dfjogos.replace(np.nan,0.0)
dfeventos = dfeventos.replace(np.nan,0.0)
```

#### TRATAMENTO

###### coluna com time vitorioso

```python
vitA = dfjogos['intHomeScore'] < dfjogos['intAwayScore'] #vitoria time de fora
vitH = dfjogos['intHomeScore'] > dfjogos['intAwayScore'] #vitoria time da casa
VH = dfjogos.strHomeTeam[vitH]  
VA = dfjogos.strAwayTeam[vitA] 
dfjogos['vitorioso'] = VH
```

```python
dfjogos['vitorioso'].fillna(value = VA, inplace = True )
#dfjogos['vitorioso'].fillna(value = 'empate', inplace = True )
```

```python
dfjogos
```

###### coluna com time derrotado

```python
derrA = dfjogos['intHomeScore'] > dfjogos['intAwayScore']
derrH = dfjogos['intHomeScore'] < dfjogos['intAwayScore']
DH = dfjogos.strHomeTeam[derrH]  
DA = dfjogos.strAwayTeam[derrA] 
dfjogos['derrotado'] = DH
```

```python
dfjogos['derrotado'].fillna(value = DA, inplace = True )
#dfjogos['derrotado'].fillna(value = 'empate', inplace = True )
```

###### teste empates

```python
emp = dfjogos['intHomeScore'] == dfjogos['intAwayScore']
EMP = dfjogos.strHomeTeam[emp]
dfjogos['empate'] = EMP
# A coluna empate traz o nomo do time da casa que empatou no caso de vitoria de qualquer lado NaN.
```
